```python
import matplotlib.pyplot as plt

font = {
    'family': 'serif',
    'color':  'black',
    'weight': 'normal',
    'size': 16,
}

file = open('thermalization_m100_w1_Nt10_tau02718.dat', 'r')
data = file.readlines()
Ns = [int(record.split()[0]) for record in data]
x2 = [float(record.split()[1]) for record in data]
sx2 = [float(record.split()[2]) for record in data]

plt.errorbar(Ns, x2, yerr=sx2, fmt='ko', label='Monte-Carlo method')

plt.title(r'$m = 100$, $\omega = 1$, $N_t = 10$, $\tau = 0.2718$, $N_p = 1000$')
plt.ylabel(r'$\langle q^2 \rangle$')
plt.xlabel('Number of sweeps')

plt.legend()
plt.savefig('thermalization.pdf', bbox_inches = 'tight')
plt.show()
```

```python
import matplotlib.pyplot as plt

font = {
    'family': 'serif',
    'color':  'black',
    'weight': 'normal',
    'size': 16,
}

file = open('energy_m100_w1_Nt10_tau02718.dat', 'r')
data = file.readlines()

Ns = [int(record.split()[0]) for record in data]
k = [float(record.split()[1]) for record in data]
sk = [float(record.split()[2]) for record in data]
v = [float(record.split()[3]) for record in data]
sv = [float(record.split()[4]) for record in data]
e = [float(record.split()[5]) for record in data]
se = [float(record.split()[6]) for record in data]

plt.errorbar(Ns, k, yerr=sk, fmt='ro', label='Kinetic energy')
plt.errorbar(Ns, v, yerr=sv, fmt='bo', label='Potential energy')
plt.errorbar(Ns, e, yerr=se, fmt='ko', label='Full energy')

plt.title(r'$m = 100$, $\omega = 1$, $N_t = 10$, $\tau = 0.2718$, $N_p = 1000$')
plt.ylabel('Energy')
plt.xlabel('Number of sweeps')

plt.legend()
plt.savefig('thermalization.pdf', bbox_inches = 'tight')
plt.show()
```
