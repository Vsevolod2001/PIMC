```python
import numpy as np
import matplotlib.pyplot as plt
import time
n=6
M=[10**i for i in range(1,n+1)]
T1=[0]*(n)
T2=[0]*(n)
T3=[0]*(n)
def plus_one(x):
    return x+1
V_plus_one=np.vectorize(plus_one)
for i in range(n):
    N=M[i]
    X=np.random.random(N)
   
    t1start=time.time()
    for j in range(N):
        X[j]=X[j]+1
    t1end=time.time()
    T1[i]=t1end-t1start
    
    t2start=time.time()
    X=list(map(lambda x:x+1,X))
    t2end=time.time()
    T2[i]=t2end-t2start
    
    t3start=time.time()
    X=V_plus_one(X)
    t3end=time.time()
    t3=t3end-t3start
    T3[i]=t3
    
    print(i+1)
    del(X)
plt.figure()
plt.plot(M,T1)
plt.plot(M,T2)
plt.plot(M,T3)
plt.xlabel("N")
plt.ylabel("t, c")
plt.grid(True)
plt.title(r"$\frac{T_{for}}{T_{map}}=$"+format((T1[n-1]/T2[n-1]),".2f")+r" $\frac{T_{for}}{T_{vectorize}}=$"+format(T1[n-1]/T3[n-1],".2f"))
plt.legend(["for","map","np.vectorize"])
plt.show()
del(T1)
del(T2)
del(T3)
```
