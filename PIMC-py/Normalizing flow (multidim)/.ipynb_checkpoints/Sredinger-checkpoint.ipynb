```python
import numpy as np
import matplotlib.pyplot as plt
L=10
h1=0.005
N=int(2*L/h1)
H=np.zeros((N,N))
H[0][0]=1
H[N-1][N-1]=1

for i in range(1,N-1):
    H[i][i-1]=-0.5/h1**2
    H[i][i]=(1/h1**2)+V(-L+i*h1)
    H[i][i+1]=-0.5/h1**2
E=np.linalg.eigh(H)
v=np.transpose(E[1])
X=[-L+i*h1 for i in range(N)]
fig=plt.figure()
plt.axis([-L,L,-1,1])
plt.scatter(X,(v[2])**2/h1,s=1)
plt.show()
```

```python
class Solver:
    def __init__(self,L1,L2,h,V):
        self.L1=L1
        self.L2=L2
        self.h=h
        self.N=int((L2-L1) / h)
        self.V=V
        self.x=np.arange(L1,L2,h)
        self.H=np.zeros((self.N,self.N))
        self.values=np.zeros((self.N))
        self.vectors=np.zeros((self.N,self.N))
    
    def compute_H(self):
        D = self.V(self.x)+(1/self.h**2)
        d = (-0.5/self.h**2)*np.ones((self.N-1))
        self.H = np.diag(D)+np.diag(d,-1)+np.diag(d,1)
        return self
    
    def compute_eigen(self):
        E = np.linalg.eigh(self.H)
        self.values = E[0]
        self.vectors = np.transpose(E[1])/(self.h)**0.5
        return self
    
    def compute_all(self):
        self.compute_H().compute_eigen()
        return self
    
    def mean(self,f,n):
        F = f(self.x)
        psi = self.vectors[n]
        return self.h*np.dot(psi**2,F)
    
    def plot(self,n):
        fig=plt.figure()
        plt.grid(True)
        plt.plot(self.x,self.vectors[n])
        plt.show()
    

        
```

```python
def V_osc(x):
    return x**2/2
solv=Solver(-20,20,0.01,V_osc)
solv.compute_all()
solv.plot(0)
print(Solv.mean(V_osc,0))
```
