Consider the quantum system with Hamiltonian
$$H=\frac{p^2}{2}+V(x) \qquad [x,p]=i $$
Suppose there exists $f(x)$ such that
$$H=\frac{f(x)-ip}{\sqrt{2}}\frac{f(x)+ip}{\sqrt{2}}+\varepsilon_0$$
Comparing the both expressions we get that
$$f'(x)=f^2(x)-2(V-\varepsilon_0)$$
Suppose there exists $\psi(x)\in L_2(\mathbb{R})$:
$$\frac{f(x)+ip}{\sqrt{2}} \psi(x)=0$$
Then $\varepsilon_0$ и $\psi$- energy and wave function of the ground state.
$$\psi'=-f(x)\psi$$
$$\psi=N\exp\left(-\int_{x_0}^x f(s) ds\right)$$
Average kinetic energy:
$$<K>=\int_{\mathbb{R}}\psi \frac{p^2}{2} \psi dx=-\frac12\int_{\mathbb{R}}\psi \psi'' dx=\frac12\int_{\mathbb{R}}\psi \psi'' dx=\int_{\mathbb{R}}\frac{f^2}{2}\psi^2 dx$$
Average potential energy:
$$<V>=\int_{\mathbb{R}} V(x)\psi^2(x) dx=\int_{\mathbb{R}}\left(\varepsilon_0-\frac{f'-f^2}{2}\right)\psi^2(x) dx$$
Integration by paths:
$$\int_{\mathbb{R}} f'\psi^2 dx=-2\int_{\mathbb{R}} f\psi'\psi dx=2\int_{\mathbb{R}} f^2\psi^2 dx$$
$$<V>=\varepsilon_0-\int_{\mathbb{R}} \frac{f^2}{2} \psi^2 dx$$
$$<K>=<\frac{f^2}{2}> \qquad <V>=\varepsilon_0-<\frac{f^2}{2}>$$

    $$E=\hbar \omega \quad L=\sqrt{\frac{\hbar}{m\omega}} \quad \Pi=\sqrt{\hbar m \omega}$$

```python
import numpy as np
from scipy.special import kn

C=0.5772156649
def my_K1_small(z,M=1000):
    a=z/2
    b=0
    psi=-C
    k=1/z
    for m in range(M):
        b=(np.log(z/2)-1/(2*m+2)-psi)/(m+1)
        k+=a*b
        
        psi+=1/(m+1)
        a*=(z/2) ** 2
        a*=1/(m+1) ** 2
    return k   
def my_K1_big(z,M=1000):
    b = 1
    k=0
    for m in range(M):
        k += b * (2*z) ** (-m)
        b *= (1-(2*m+1)/4)/(m+1)
    k*=(np.pi/(2*z))**0.5 * np.exp(-z)    
    return k 

def my_K1(z,M=1000,z0=10.5):
    if z<=z0:
        return my_K1_small(z,M)
    else:
        return my_K1_big(z,10*M)

print(my_K1_small(2)-kn(1, 2))       
    
```
