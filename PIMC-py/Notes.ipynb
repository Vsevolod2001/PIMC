Consider the quantum system with Hamiltonian
$$H=\frac{p^2}{2}+V(x) \qquad [x,p]=i $$
Suppose there exists $f(x)$ such that
$$H=\frac{f(x)-ip}{\sqrt{2}}\frac{f(x)+ip}{\sqrt{2}}+\varepsilon_0$$
Comparing the both expressions we get that
$$f'(x)=f^2(x)-2(V-\varepsilon_0)$$
Suppose there exists $\psi(x)\in L_2(\mathbb{R})$:
$$\frac{f(x)+ip}{\sqrt{2}} \psi(x)=0$$
Then $\varepsilon_0$ и $\psi$- energy and wave function of the ground state.
$$\psi'=-f(x)\psi$$
$$\psi=N\exp\left(-\int_{x_0}^x f(s) ds\right)$$
Average kinetic energy:
$$<K>=\int_{\mathbb{R}}\psi \frac{p^2}{2} \psi dx=-\frac12\int_{\mathbb{R}}\psi \psi'' dx=\frac12\int_{\mathbb{R}}\psi \psi'' dx=\int_{\mathbb{R}}\frac{f^2}{2}\psi^2 dx$$
Average potential energy:
$$<V>=\int_{\mathbb{R}} V(x)\psi^2(x) dx=\int_{\mathbb{R}}\left(\varepsilon_0-\frac{f'-f^2}{2}\right)\psi^2(x) dx$$
Integration by paths:
$$\int_{\mathbb{R}} f'\psi^2 dx=-2\int_{\mathbb{R}} f\psi'\psi dx=2\int_{\mathbb{R}} f^2\psi^2 dx$$
$$<V>=\varepsilon_0-\int_{\mathbb{R}} \frac{f^2}{2} \psi^2 dx$$
$$<K>=<\frac{f^2}{2}> \qquad <V>=\varepsilon_0-<\frac{f^2}{2}>$$

    $$E=\hbar \omega \quad L=\sqrt{\frac{\hbar}{m\omega}} \quad \Pi=\sqrt{\hbar m \omega}$$

Morse potential:
$$H=\frac{P^2}{2m}+\frac{g^2}{2}\left[(e^{-qX}-1)^2-1\right] \qquad q>0$$
By analogy with the oscillator, we introduce dimensionless variables using the following characteristic scales:
$$E=g^2 \qquad L=\frac{\hbar}{\sqrt{m}g} \qquad \Pi=\sqrt{m}g$$
Hamiltonian takes the form:
$$H=\frac{p^2}{2}+\frac12\left[(e^{-\alpha x}-1)^2-1\right]$$
where
$$\alpha=qL=\frac{q\hbar}{g\sqrt{m}}>0$$
is the dimensionless parametr.
$$f(x)=-e^{-\alpha x}+\left(1-\frac{\alpha}{2}\right)x \qquad \varepsilon_0=-\frac{(2-\alpha)^2}{8}$$
$$\psi=N\exp\left[-\frac{1}{\alpha}e^{-\alpha x}-\frac{2-\alpha}{2}x\right]$$
Note, that bound states are exist only if $0<\alpha<2$

Average value of arbitrary function $\varphi(x)$ is:
$$<\varphi(x)>=\int_{\mathbb{R}}\varphi(x) \psi^2(x) dx=\int_{\mathbb{R}}\varphi(x) N^2\exp\left[-\frac{2}{\alpha}e^{-\alpha x}-(2-\alpha)x\right]dx$$
We substitute the variable:
$$y=\frac{2}{\alpha}e^{-\alpha x} \qquad x=-\frac{1}{\alpha}\ln\left(\frac{\alpha}{2}y\right)$$
$$<\varphi(x)>=\frac{N^2}{2}\left(\frac{\alpha}{2}\right)^{\frac{2}{\alpha}-2}\int_{\mathbb{R}^{+}}\varphi\left(-\frac{1}{\alpha}\ln\left(\frac{\alpha}{2}y\right)\right)e^{-y}y^{\frac{2-\alpha}{\alpha}-1} dy$$
We calculate the normalizing factor, using the condition $<1>=1$:
$$N^{-2}=\frac{1}{2}\left(\frac{\alpha}{2}\right)^{\frac{2}{\alpha}-2}\Gamma\left(\frac{2-\alpha}{\alpha}\right)$$
The main results: 
$$<\varphi(x)>=\int_{\mathbb{R}^{+}}\varphi\left(-\frac{1}{\alpha}\ln\left(\frac{\alpha}{2}y\right)\right)\frac{e^{-y}y^{\frac{2-\alpha}{\alpha}-1}}{\Gamma\left(\frac{2-\alpha}{\alpha}\right)}dy$$
$$<K>=<\frac{f^2(x)}{2}>=\frac{\alpha}{4}\left(1-\frac{\alpha}{2}\right)$$

Consider the model of scalar field with (Euclidean) action:
$$\tilde A(\varphi)=\frac{1}{2}\int\left\{\dot\varphi^2-q\lambda e^{q\varphi}+\lambda^2e^{2q\varphi}\right\}dt$$
$$q<0 \qquad \lambda<0$$
There are initional condition for the field $\varphi$:
$$\varphi(0)=-\infty$$
Transform the action to the form:
$$\tilde A(\varphi)=\frac{1}{2}\int\left\{\dot\varphi^2-\lambda^2\left[\left(e^{q\varphi}-\frac{q}{2\lambda}\right)^2-\left(\frac{q}{2\lambda}\right)^2\right]\right\}dt$$
Transform the action by the following linear substitutions (the initial condition is invariant to these substitutions):

$$\varphi \to \theta_1 \varphi+\theta_2 \qquad t \to \theta_3 t$$

$$\tilde A(\varphi)=\frac{1}{2}\int\left\{\frac{\theta_1^2}{\theta_3}\dot\varphi^2-\lambda^2\theta_3e^{2q\theta_2}\left[\left(e^{q\theta_1\varphi}-\frac{qe^{-q\theta_2}}{2\lambda}\right)^2-\left(\frac{qe^{-q\theta_2}}{2\lambda}\right)^2\right]\right\}dt$$

Сhoose the parameters equal to
$$\theta_1^2=\theta_3 \qquad e^{q\theta_2}=\frac{q}{2\lambda}$$
$$\tilde A(\varphi)=\frac{1}{2}\int\left\{ \dot \varphi^2+\lambda^2\left(\frac{q}{2\lambda}\right)^2 \theta_3\left[(e^{q\sqrt{\theta_3} \varphi}-1)^2-1\right]\right\}dt$$
$$\theta_3=\frac{4}{q^2} \qquad \sqrt{\theta_3}=\frac{2}{|q|} \qquad q\sqrt{\theta_3}=-2$$
The final form:
$$\tilde A(\varphi)=\frac{1}{2}\int\left\{\dot \varphi^2+\left[(e^{-2\varphi}-1)^2-1\right]\right\}dt$$
So, this model is equivalent to previosly considered Morse potential with the parameter $\alpha=2$ - there is no one bound state in this model!

The simpliest example-harmonic oscillator

$$\tilde H=\frac{P^2}{2m}+\frac{m\omega^2 X^2}{2} \qquad [X,P]=i\hbar$$
We introduce dimensionless variables:
    $$\tilde H=E H \quad X=Lx \quad P=\Pi p$$
    $$E=\hbar \omega \quad L=\sqrt{\frac{\hbar}{m\omega}} \quad \Pi=\sqrt{\hbar m \omega}$$
And we transform our problem to the previously considered type 
$$H=\frac{p^2}{2}+\frac{x^2}{2}$$

Relativistic oscillator
$$\tilde H=\sqrt{P^2c^2+m^2c^4}+\frac{m\omega^2X^2}{2} \qquad [X,P]=i\hbar$$
There are four dimension variables: $\hbar, m, c, \omega$

Case1
basic scales: $\hbar, m, \omega$
$$\Pi=\sqrt{\hbar m\omega} \qquad L=\sqrt{\frac{\hbar}{m\omega}} \qquad E=\hbar \omega$$
$$\hbar \omega H=\sqrt{p^2\hbar m \omega c^2+m^2c^4}+\frac{1}{2}\frac{\hbar}{ m \omega}m\omega^2 x^2 $$
$$H=\frac{1}{\sigma}\sqrt{p^2\sigma+1}+\frac{x^2}{2} \qquad \sigma=\frac{\hbar \omega}{mc^2}$$
non-relativistic:
$$\sigma<<1 \qquad H=\frac{p^2}{2}+\frac{x^2}{2}+\frac{1}{\sigma}$$
$$\sigma>>1 \qquad H=\frac{|p|}{\sqrt{\sigma}}+\frac{x^2}{2}$$

Case2:
basic scales: $\hbar, \omega, c$
$$L=\frac{c}{\omega} \qquad \Pi=\frac{\hbar\omega}{c} \qquad E=\hbar \omega$$
$$\hbar \omega H=\sqrt{p^2c^2\hbar^2\omega^2/c^2+m^2c^4}+\frac{m\omega^2x^2}{2}\frac{c^2}{\omega^2}$$
$$H=\frac{1}{\sigma}\sqrt{p^2\sigma^2+1}+\frac{x^2}{2\sigma} \qquad \sigma=\frac{\hbar \omega}{mc^2}$$
non-relativistic:
$$\sigma<<1 \qquad H=\frac{p^2\sigma}{2}+\frac{x^2}{2\sigma}$$
relativistic:
$$\sigma>>1 \qquad H=|p|+\frac{x^2}{2\sigma}$$
